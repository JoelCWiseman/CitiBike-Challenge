```python
# Path to the cleaned data CSV file
cleaned_data_path = r'C:\Users\joelw\GWAnalyticsClass\Challenges\Challenge18\2022_all_data.csv'

# Read the cleaned data CSV into a DataFrame
cleaned_data_df = pd.read_csv(cleaned_data_path)

# Splitting "started_at" into "started_date" and "started_time"
cleaned_data_df[['started_date', 'started_time']] = cleaned_data_df['started_at'].str.split(' ', expand=True)

# Splitting "ended_at" into "ended_date" and "ended_time"
cleaned_data_df[['ended_date', 'ended_time']] = cleaned_data_df['ended_at'].str.split(' ', expand=True)

# Renaming columns
cleaned_data_df.rename(columns={'started_at': 'started_datetime', 'ended_at': 'ended_datetime'}, inplace=True)

# Print the first few rows of the DataFrame to verify the changes
cleaned_data_df.head()
```

```python
# Drop the unnecessary columns
cleaned_data_df.drop(['started_datetime', 'ended_datetime'], axis=1, inplace=True)

cleaned_data_df.head()
```

```python
# Calculate ride duration in minutes
time_format = "%H:%M:%S"
cleaned_data_df['ride_duration'] = (pd.to_datetime(cleaned_data_df['ended_time'], format=time_format) - pd.to_datetime(cleaned_data_df['started_time'], format=time_format)).dt.total_seconds() / 60

cleaned_data_df.head()
```

```python
# Calculate distance using Haversine formula (in miles)
def haversine_distance(lat1, lon1, lat2, lon2):
    R = 3958.8  # Earth radius in miles
    
    lat1_rad = np.radians(lat1)
    lon1_rad = np.radians(lon1)
    lat2_rad = np.radians(lat2)
    lon2_rad = np.radians(lon2)
    
    dlat = lat2_rad - lat1_rad
    dlon = lon2_rad - lon1_rad
    
    a = np.sin(dlat / 2)**2 + np.cos(lat1_rad) * np.cos(lat2_rad) * np.sin(dlon / 2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
    
    distance = R * c
    return distance

cleaned_data_df['distance_miles'] = haversine_distance(
    cleaned_data_df['start_lat'], cleaned_data_df['start_lng'],
    cleaned_data_df['end_lat'], cleaned_data_df['end_lng']
)

# Add a new column for approximate distance traveled
def categorize_distance(distance):
    if distance < 0.5:
        return "0-0.49 mile"
    elif distance < 1:
        return "0.5-0.99 mile"
    elif distance < 1.5:
        return "1-1.49 miles"
    elif distance < 2:
        return "1.5-1.99 miles"
    elif distance < 2.5:
        return "2-2.49 miles"
    elif distance < 3:
        return "2.5-2.99 miles"
    elif distance < 3.5:
        return "3-3.49 miles"
    elif distance < 4:
        return "3.5-3.99 miles"
    elif distance < 10:
        return "4-9.99 miles"
    else:
        return "10+ miles"

cleaned_data_df['approximate_distance_traveled'] = cleaned_data_df['distance_miles'].apply(categorize_distance)

cleaned_data_df.to_csv('cleaned_data_2022.csv', index=False)

cleaned_data_df.head()
```

```python
# Count the occurrences of each start location
top_start_locations = cleaned_data_df[['start_lat', 'start_lng', 'start_station_name']].value_counts().reset_index(name='count')

# Select the top 50 start locations
top_50_start = top_start_locations.head(50)

# Save the new DataFrame to a CSV
top_50_start.to_csv('top_50_start.csv', index=False)

top_50_start.head()
```

```python
# Count the occurrences of each end location
top_end_locations = cleaned_data_df[['end_lat', 'end_lng', 'end_station_name']].value_counts().reset_index(name='count')

# Select the top 50 end locations
top_50_end = top_end_locations.head(50)

# Save the new DataFrame to a CSV
top_50_end.to_csv('top_50_end.csv', index=False)

top_50_end.head()
```

```python
# Group by start and end locations and count occurrences
top_combinations = cleaned_data_df.groupby(['start_lat', 'start_lng', 'start_station_name', 'end_lat', 'end_lng', 'end_station_name']).size().reset_index(name='count')

# Sort by count in descending order
top_combinations = top_combinations.sort_values(by='count', ascending=False)

# Select the top 50 combinations
top_50_combinations = top_combinations.head(50)

# Save the new DataFrame to a CSV
top_50_combinations.to_csv('top_50_combinations.csv', index=False)

top_50_combinations.head()
```
